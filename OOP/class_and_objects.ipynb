- Python is an object oriented programming language
- Unlike procedure-oriented programming, where the main emphasis is on functions, object-oriented programming stresses on objects.
- An object is simply a collection of data (variables) and methods (functions) that act on those data.
- similarly, al class is a blueprint for that object.
- Like function definitions begin with the <b>def</b> keyword in Python, class definitions begin with a <b>class</b> keyword.
- The first string inside the class is called docstring and has a brief description of the class.
- Although not mandatory, this is highly recommented.

## Create a class

```python
class Data:
    data = 1.69
print(data)  # can't directly access the data variable needs to create an objcets of the class
```

```python
class Data:
    data = 1.69

print(Data)  # <class '__main__.Data'>
print(data) 
```

```python
class Data:
    data = 1.69

obj = Data()
print(obj.data)  # accessible prints : 1.69
```

## Function init()

```python
class Data:
    def __init__(self, name, age, yob) :
        self.name = name
        self.age = age
        self.yob = yob

obj = Data("naim", 20, 2002)
print(obj.name)
print(obj.age)
print(obj.yob)
```

## Methods

```python
# Creating multiple objects
# Methods in a class

class Data:
    def __init__(self, name, department) :
        self.name = name
        self.department = department
    
    def info(self) :
        print('hi! I\'m ' + self.name)
        print('Department : ' + self.department)

naim = Data("naim", "CSE")
ashraful = Data("Ashraful", "CSE")
naim.info()
ashraful.info()
```

## Self parameter
- The self parameter is a reference to the current instance of the class, and is used to access variables that belongs to the class.
- It does not have to be named self, another names are allowed.


```python
# no use of self parameter

'''The following code are made without using self operator.
self operator replaced with this operator
'''

class Person :
    def __init__(this, name, age):
        this.name = name
        this.age = age
    def printInfo(this) :
        print("name : " + this.name)
        print("age : " + str(this.age))
person = Person('abdul alim', 24)
person.printInfo()
```

## Real time class example
### Creating a class to draw a Rectangle

```python
# Creating a class to draw a rectangle

class Rectangle(object) :
    # constructor
    def __init__(this, width, height, color) :
        this.width = width
        this.height = height
        this.color = color
    # Method
    def drawRectangle(this) :
        plt.gca().add_patch(plt.Rectangle((0, 0), this.width, this.height, fc=this.color))
        plt.axis('scaled')
        plt.show()

    
# import library to draw the rectangle

import matplotlib.pyplot as plt
%matplotlib inline

# Creating an object blue rectangle
first_rectangle = Rectangle(2, 1, 'blue')
first_rectangle.drawRectangle()
print(first_rectangle.height)
print(first_rectangle.width)
print(first_rectangle.color)

# Learning the methods that can be utilized on the object 'first_rectangle'
print(dir(first_rectangle))


# we can change the properties of the rectangle
first_rectangle.color = 'green'
first_rectangle.height = 2
first_rectangle.drawRectangle()

second_rectangle = Rectangle(1.25, 0.6, 'red')
second_rectangle.drawRectangle()
```

# Creating a class to draw a circle

```python
class Circle(object) :
    
    # constructor
    def __init__(this, radius, color) :
        this.radius = radius
        this.color = color
    
    # method
    def increase_rad(this, r) :
        this.radius += r
    # method
    def drawCircle(this) :
        plt.gca().add_patch(plt.Circle((0, 0), this.radius, fc=this.color))
        plt.axis("scaled")
        plt.show()
# import library
import matplotlib.pyplot as plt

# Creatig an object of Circle color blue
first_circle = Circle(3.14, 'blue')
first_circle.drawCircle()

first_circle.increase_rad(0.86)
first_circle.drawCircle()
first_circle.color = 'yellow'
first_circle.drawCircle()
```
